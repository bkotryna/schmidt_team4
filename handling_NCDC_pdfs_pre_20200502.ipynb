### Download all pdf's from the NCDC website

```python
# web scraping to download pdf files
# get html of the website where all the pdf files can be doanloaded from

from bs4 import BeautifulSoup
import requests


url="https://ncdc.gov.ng/diseases/sitreps/?cat=14&name=An%20update%20of%20COVID-19%20outbreak%20in%20Nigeria"
raw_html = requests.get(url).content
raw_html

parsed_html = BeautifulSoup(raw_html)
parsed_html
```

```python
# get all links to pdfs into a list
# also make a list with pdf file names

table_with_pdfs = parsed_html.find("table", {"class": "table table-striped"})

links = []
file_names = []

for element in table_with_pdfs.find_all("a"):
    link = 'https://ncdc.gov.ng' + element.attrs["href"]
    links.append(link)
    
    file_name = element.attrs['download']
    file_names.append(file_name)
```

```python
# loop through all links, download pdf files and name them in a nice way (date + week number)

for i, item in enumerate(links):
    pdf_content = requests.get(item).content

    filename = 'data/NCDC_pdfs/' + file_names[i].split('Nigeria_')[1]
    new_name = 
    with open(pdf_name, 'wb') as f:
        f.write(pdf_content)
```

```python
# more targeted approach to fix individual issues
# try on all pdfs (apart from July 7)

# => works on all files apart from July 7 => yay!!
# save the resulting df as .csv

import os

base_path = '/mnt/d/Katrytes/Rasliava/Coding_Statistics/Python/Courses/2020-07_Data_for_Science/Hackathlon/'
extra_path = 'schmidt_team4/data/NCDC_pdfs/all_good/'
full_path = base_path + extra_path
os.chdir(full_path)
print(os.getcwd())

import pdfplumber
from glob import glob
import pandas as pd
import re

# get pdf file names
files = glob('*.pdf')

# sort alphabeticaly
files = sorted(files)

# set up a df where the output will go
column_names = ['state', 'total_confirmed', 'new_confirmed', 'total_discharged', 'new_dicharged',
               'total_deaths', 'new_deaths', 'total_active', 'date']
pdf_df = pd.DataFrame(columns=column_names)

# loop through all files
for file in files:
    
    # just to keep track
    print(file)
    
    # extract date from the title
    pdf_date = file.split('/')[-1].split('_')[0]
    # print(pdf_date)
    
    # read pdf file, extract text from the page with the table
    pdf = pdfplumber.open(file)
    table_page = pdf.pages[1]
    
    
    # split table page into lines
    page_text = table_page.extract_text()
    page_lines = page_text.split('\n')
    # print(page_lines)
    
    
    # establish where the table text starts & finishes
    # then save table text into a new variable

    counter_start = 0
    counter_stop = 0
    for i, line in enumerate(page_lines):
        if 'CASE' in line:
            counter_start = i + 1
        if 'Total' in line or 'TOTAL' in line:
            counter_stop = i

    # print (counter_start, counter_stop)
    table_text = page_lines[counter_start:counter_stop]
    table_text[:] = [x for x in table_text if x.strip()]
    # print(table_text)

    # convert text from the table into a pandas df

    
    for number, line in enumerate(table_text):
        
        # troubleshooting extras
        
        # when state name is split over two lines => merge them
        if len(line) < 16:
            line = line + ' ' + table_text[number+1]
          
        # when there's an NA => skip line
        # this may seem radical, but there aren't many (and, key point, they throw errors otherwise)
        if 'N/A' in line:
            continue
        
        # get text, ie state name
        state_name = "".join(re.split("[^a-z A-Z]*", line)).strip()
        rest = line[len(state_name):]
        
        words = rest.split()
        words.insert(0, state_name)
        line_dict = {}
        line_dict['date'] = pdf_date
        for i, column in enumerate(column_names[:-1]): # ignore the 'date' column
            # print(words[i])
            line_dict[column] = words[i]
        pdf_df = pdf_df.append(line_dict, ignore_index=True)
        
all_good_df = pdf_df.copy()
all_good_df


# save the resulting df as .csv

import os


base_path = '/mnt/d/Katrytes/Rasliava/Coding_Statistics/Python/Courses/2020-07_Data_for_Science/Hackathlon/'
extra_path = 'schmidt_team4/data/NCDC_pdfs/output_csv/'
full_path = base_path + extra_path
os.chdir(full_path)
print(os.getcwd())

all_good_df.to_csv('all_good.csv')
```
